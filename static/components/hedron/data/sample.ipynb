```python
from mpl_toolkits.mplot3d import Axes3D

solid =  [[
      -1.3333333333333333,
      -0.7698003589195005,
      -1.088662107903635
    ],
    [
      -0.8333333333333338,
      -1.0584754935143137,
      -0.27216552697590907
    ],
    [
      -0.5000000000000002,
      -0.28867513459481264,
      -0.8164965809277261
    ],
    [
      -0.444444444444445,
      -1.21885056828921,
      -1.1793839502289383
    ]
]
import matplotlib.pyplot as plt


# Extract x, y, z coordinates
x = [point[0] for point in solid]
y = [point[1] for point in solid]
z = [point[2] for point in solid]

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='b', marker='o')
# Add lines connecting the points
lines = list(zip(solid, solid[1:] + [solid[0]]))  # Connect each point to the next, and the last to the first
for line in lines:
    x_line = [line[0][0], line[1][0]]
    y_line = [line[0][1], line[1][1]]
    z_line = [line[0][2], line[1][2]]
    ax.plot(x_line, y_line, z_line, c='r')

# Label axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
